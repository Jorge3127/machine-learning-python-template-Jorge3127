```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axis = plt.subplots(2, 3, figsize=(18, 12))  

# Primer fila de grÃ¡ficos
sns.countplot(ax=axis[0, 0], data=df, x="price", palette='Blues', hue="price")
axis[0, 0].legend(title="PRECIOS", loc= "upper right", fontsize= "small", title_fontsize="medium")
axis[0, 0].set_title("DistribuciÃ³n de Precios")

sns.countplot(ax=axis[0, 1], data=df, x="minimum_nights", palette='Blues', hue="minimum_nights")
axis[0, 1].legend(title="NOCHES MÃNIMAS")
axis[0, 1].set_title("DistribuciÃ³n de Noches MÃ­nimas")

sns.countplot(ax=axis[0, 2], data=df, x="room_type", palette='Blues', hue="room_type")
axis[0, 2].legend(title="TIPO DE HABITACIÃ“N")
axis[0, 2].set_title("DistribuciÃ³n de Tipos de HabitaciÃ³n")

# Segunda fila de grÃ¡ficos
sns.countplot(ax=axis[1, 0], data=df, x="neighbourhood_group", palette='Blues', hue="neighbourhood_group")
axis[1, 0].legend(title="GRUPO DE VECINDARIO")
axis[1, 0].set_title("DistribuciÃ³n por Grupo de Vecindario")

sns.countplot(ax=axis[1, 1], data=df, x="neighbourhood", palette='Blues', hue="neighbourhood")
axis[1, 1].legend(title="VECINDARIO")
axis[1, 1].set_title("DistribuciÃ³n por Vecindario")

sns.countplot(ax= axis[1, 2], data=df, x="number_of_reviews", palette='Blues', hue="number_of_reviews")
axis[1, 2].legend(title= "NUMERO DE RESEÃ‘AS")
axis[1, 2].set_title("DistribuciÃ³n por ReseÃ±as")

plt.tight_layout()
plt.show()

```

```python
fig, axis = plt.subplots(2, 2, figsize = (10, 6))

print("PRIMERO REALIZAREMOS UN ANALISIS NUMERICO - NUMERICO\n")

print("1. REPRESENTAREMOS LA RELACION ENTRE EL PRECIO CON EL MINIMO DE NOCHES Y EL NUMERO DE RESEÃ‘AS")
# Crear un diagrama de dispersiÃ³n mÃºltiple
sns.regplot(ax = axis[0, 0], data = df, x = "minimum_nights", y = "price")
sns.heatmap(df[["price", "minimum_nights"]].corr(), annot = True, fmt = ".2f", ax = axis[1, 0], cbar = False)

sns.regplot(ax = axis[0, 1], data = df, x = "number_of_reviews", y = "price").set(ylabel=None)
sns.heatmap(df[["price", "number_of_reviews"]].corr(), annot = True, fmt = ".2f", ax = axis[1, 1])

# Ajustar el layout
plt.tight_layout()

# Mostrar el plot
plt.show()
```

```python
fig, axis = plt.subplots(2, 2, figsize = (10, 6))

print("2. REPRESENTAREMOS LA RELACION ENTRE EL PRECIO CON EL NUMERO DE  ANFITRIONES Y LA DISPONIBILIDAD")
# Crear un diagrama de dispersiÃ³n mÃºltiple

sns.regplot(ax = axis[0, 0], data = df, x = "calculated_host_listings_count", y = "price").set(ylabel=None)
sns.heatmap(df[["price", "calculated_host_listings_count"]].corr(), annot = True, fmt = ".2f", ax = axis[1, 0])

sns.regplot(ax = axis[0, 1], data = df, x = "availability_365", y = "price").set(ylabel=None)
sns.heatmap(df[["price", "availability_365"]].corr(), annot = True, fmt = ".2f", ax = axis[1, 1])

# Ajustar el layout
plt.tight_layout()

# Mostrar el plot
plt.show()
```

```python
fig, axis = plt.subplots(1, 3, figsize = (15, 6))

print("REPRESENTAREMOS LAS VARIABLES CATEGORICAS UNA A UNA\n")

sns.countplot(ax=axis[0], data=df, x="last_review", hue="room_type").set(ylabel=None)
axis[0].set_xticklabels(axis[0].get_xticklabels(), rotation=45, ha='right')
axis[0].set_xlabel('FECHA ULTIMA RESEÃ‘A')
axis[0].set_title('TIPOS DE HABITACIONES POR ULTIMA RESEÃ‘A')

sns.countplot(ax=axis[1], data=df, x="neighbourhood", hue="room_type").set(ylabel=None)
axis[1].set_xticklabels(axis[1].get_xticklabels(), rotation=45, ha='right')
axis[1].set_xlabel('Neighbourhood')
axis[1].set_title('Count of Room Types by Neighbourhood')

sns.countplot(ax=axis[2], data=df, x="neighbourhood_group", hue="room_type").set(ylabel=None)
axis[2].set_xticklabels(axis[2].get_xticklabels(), rotation=45, ha='right')
axis[2].set_xlabel('Neighbourhood Group')
axis[2].set_title('Count of Room Types by Neighbourhood Group')

plt.tight_layout()
plt.show(block=True)

```

```python
fig, axis = plt.subplots(1, 3, figsize = (10, 6))

crosstab_last_review= pd.crosstab(df['last_review'], df['room_type'])
crosstab_last_review.div(crosstab_last_review.sum(1).astype(float), axis=0)
crosstab_last_review.plot(kind='bar', stacked=True, ax = axis[0])

crosstab_neighbourhood = pd.crosstab(df['neighbourhood'], df['room_type'])
crosstab_neighbourhood.div(crosstab_neighbourhood.sum(1).astype(float), axis=0)
crosstab_neighbourhood.plot(kind='bar', stacked=True, ax = axis[1])

crosstab_neighbourhood_group = pd.crosstab(df['neighbourhood_group'], df['room_type'])
crosstab_neighbourhood_group.div(crosstab_neighbourhood_group.sum(1).astype(float), axis=0)
crosstab_neighbourhood_group.plot(kind='bar', stacked=True, ax = axis[2])


plt.tight_layout()
plt.show()
```
